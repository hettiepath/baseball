```
import time
import pandas as pd
from tqdm.notebook import tqdm
import re

from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```
year = []
name = []
team = []
avg = []
game = []
pa = []
ab = []
run = []
hit = []
b2 = []
b3 = []
hr = []
tb = []
rbi = []
sac = []
sf = []
bb = []
ibb = []
hbp = []
so = []
gdp = []
slg = []
obp = []
ops = []
mh = []
risp = []
xbh = [] 
go = []
ao = []
goao = []
gw = []
bbk = []
isop = []
xr = []
gpa = []

years = [2015, 2016, 2017,2018 ,2019]
y = [34, 35, 36, 37, 38]




for j in tqdm(range(1,4),desc = 'data'):
  if j == 3:
    url = 'https://www.koreabaseball.com/Record/Player/HitterBasic/Detail1.aspx'
  if j != 3:
    url = 'https://www.koreabaseball.com/Record/Player/HitterBasic/Basic'+str(j)+'.aspx'
  

  driver.get(url)
  time.sleep(1)

  for page in tqdm(range(len(years)),desc='years'):
    driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ddlSeason_ddlSeason > option:nth-child('+str(y[page])+')').click()
    time.sleep(1)


    if j==1:
      print(j)
      for p in range(2):
        
        
        names = driver.find_elements_by_css_selector(' td:nth-child(2) > a')
        teams = driver.find_elements_by_css_selector(' td:nth-child(3)')
        avgs = driver.find_elements_by_css_selector('td.asc')
        games = driver.find_elements_by_css_selector('td:nth-child(5)')
        pas = driver.find_elements_by_css_selector('td:nth-child(6)')
        abs = driver.find_elements_by_css_selector('td:nth-child(7)')
        runs = driver.find_elements_by_css_selector('td:nth-child(8)')
        hits = driver.find_elements_by_css_selector('td:nth-child(9)')
        b2s = driver.find_elements_by_css_selector('td:nth-child(10)')
        b3s = driver.find_elements_by_css_selector('td:nth-child(11)')
        hrs = driver.find_elements_by_css_selector('td:nth-child(12)')
        tbs = driver.find_elements_by_css_selector('td:nth-child(13)')
        rbis = driver.find_elements_by_css_selector('td:nth-child(14)')
        sacs = driver.find_elements_by_css_selector('td:nth-child(15)')
        sfs = driver.find_elements_by_css_selector('td:nth-child(16)')
      

        for i in range(len(names)):
          
          year.append(years[page])
          name.append(names[i].text)
          team.append(teams[i].text)
          avg.append(float(avgs[i].text))
          game.append(int(games[i].text))
          pa.append(int(pas[i].text))
          ab.append(int(abs[i].text))
          run.append(int(runs[i].text))
          hit.append(int(hits[i].text))
          b2.append(int(b2s[i].text))
          b3.append(int(b3s[i].text))
          hr.append(int(hrs[i].text))
          tb.append(int(tbs[i].text))
          rbi.append(int(rbis[i].text))
          sac.append(int(sacs[i].text))
          sf.append(int(sfs[i].text))

        driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ucPager_btnNo2').click()
        time.sleep(1)

      
      
    if j==2:
      print(j)

      for p in range(2):
      
        bbs = driver.find_elements_by_css_selector('td:nth-child(5)')
        ibbs = driver.find_elements_by_css_selector(' td:nth-child(6)')
        hbps = driver.find_elements_by_css_selector('td:nth-child(7)')
        sos = driver.find_elements_by_css_selector('td:nth-child(8)')
        gdps = driver.find_elements_by_css_selector('td:nth-child(9)')
        slgs = driver.find_elements_by_css_selector('td:nth-child(10)')
        obps = driver.find_elements_by_css_selector('td:nth-child(11)')
        opss = driver.find_elements_by_css_selector('td:nth-child(12)')
        mhs = driver.find_elements_by_css_selector('td:nth-child(13)')
        risps = driver.find_elements_by_css_selector('td:nth-child(14)')
    

        for i in range(len(bbs)):

          bb.append(int(bbs[i].text))
          ibb.append(int(ibbs[i].text))
          hbp.append(int(hbps[i].text))
          so.append(int(sos[i].text))
          gdp.append(int(gdps[i].text))
          slg.append(float(slgs[i].text))
          obp.append(float(obps[i].text))
          ops.append(float(opss[i].text))
          mh.append(int(mhs[i].text))
          risp.append(float(risps[i].text))

        driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ucPager_btnNo2').click()
        time.sleep(1)

    if j == 3:
      print(j)
      for p in range(2):

        xbhs = driver.find_elements_by_css_selector('td:nth-child(5)')
        gos = driver.find_elements_by_css_selector('td:nth-child(6)')
        aos = driver.find_elements_by_css_selector('td:nth-child(7)')
        goaos = driver.find_elements_by_css_selector('td:nth-child(8)')
        gws = driver.find_elements_by_css_selector('td:nth-child(9)')
        bbks = driver.find_elements_by_css_selector('td:nth-child(10)')
        isops = driver.find_elements_by_css_selector('td:nth-child(11)')
        xrs = driver.find_elements_by_css_selector('td:nth-child(12)')
        gpas = driver.find_elements_by_css_selector('td:nth-child(13)')
        


        for i in range(len(xbhs)):
          xbh.append(int(xbhs[i].text))
          go.append(int(gos[i].text))
          ao.append(int(aos[i].text))
          goao.append(float(goaos[i].text))
          gw.append(int(gws[i].text))
          bbk.append(float(bbks[i].text))
          isop.append(float(isops[i].text))
          xr.append(float(xrs[i].text))
          gpa.append(float(gpas[i].text))

        driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ucPager_btnNo2').click()
        time.sleep(1)
      

          
```
