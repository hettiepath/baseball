```
import time
import pandas as pd
from tqdm.notebook import tqdm
import re

from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```
year = []
name = []
team = []
era = []
game = []
w = []
l = []
wpct = []
ip = []
ph = []
phr = []
pbb = []
phbp = []
pso = []
r = []
er = []
whip = []
qs = []
tbf = []
np = []
pavg = []
pb2 = []
pb3 = []


years = [2015, 2016, 2017,2018 ,2019]
y = [34, 35, 36, 37, 38]


for j in range(1,3):
  url = 'https://www.koreabaseball.com/Record/Player/PitcherBasic/Basic'+str(j)+'.aspx'
    
  driver.get(url)
  time.sleep(1)

  for page in tqdm(range(len(years)),desc='years'):
  
    driver.find_element_by_css_selector('#cphContents_cphContents_cphContents_ddlSeason_ddlSeason > option:nth-child('+str(y[page])+')').click()
    time.sleep(1)


    if j==1:

      names = driver.find_elements_by_css_selector(' td:nth-child(2) > a')
      teams = driver.find_elements_by_css_selector(' td:nth-child(3)')
      eras = driver.find_elements_by_css_selector('td.asc')
      games = driver.find_elements_by_css_selector('td:nth-child(5)')
      ws = driver.find_elements_by_css_selector('td:nth-child(6)')
      ls = driver.find_elements_by_css_selector('td:nth-child(7)')
      wpcts = driver.find_elements_by_css_selector('td:nth-child(10)')
      ips = driver.find_elements_by_css_selector('td:nth-child(11)')
      phs = driver.find_elements_by_css_selector('td:nth-child(12)')
      phrs = driver.find_elements_by_css_selector('td:nth-child(13)')
      pbbs= driver.find_elements_by_css_selector('td:nth-child(14)')
      hbps = driver.find_elements_by_css_selector('td:nth-child(15)')
      psos = driver.find_elements_by_css_selector('td:nth-child(16)')
      rs = driver.find_elements_by_css_selector('td:nth-child(17)')
      ers = driver.find_elements_by_css_selector('td:nth-child(18)')
      whips = driver.find_elements_by_css_selector('td:nth-child(19)')


      for i in range(len(names)):
        
        year.append(years[page])
        name.append(names[i].text)
        team.append(teams[i].text)
        era.append(float(eras[i].text))
        game.append(int(games[i].text))
        w.append(int(ws[i].text))
        l.append(int(ls[i].text))
        wpct.append(float(wpcts[i].text))
        ip.append(int(ips[i].text[0]))
        ph.append(int(phs[i].text))
        phr.append(int(phrs[i].text))
        pbb.append(int(pbbs[i].text))
        pso.append(int(psos[i].text))
        r.append(int(rs[i].text))
        er.append(int(ers[i].text))
        whip.append(float(whips[i].text))

    if j==2:

      qss = driver.find_elements_by_css_selector('td:nth-child(7)')
      tbfs = driver.find_elements_by_css_selector(' td:nth-child(9)')
      nps = driver.find_elements_by_css_selector('td:nth-child(10)')
      pavgs = driver.find_elements_by_css_selector('td:nth-child(11)')
      pb2s = driver.find_elements_by_css_selector('td:nth-child(12)')
      pb3s = driver.find_elements_by_css_selector('td:nth-child(13)')



      for i in range(len(qss)):

        qs.append(int(qss[i].text))
        tbf.append(int(tbfs[i].text))
        np.append(int(nps[i].text))
        pavg.append(float(pavgs[i].text))
        pb2.append(int(pb2s[i].text))
        pb3.append(int(pb3s[i].text))

  print(len(name))
  print(len(pb3))
        
```
